### Build a df of the HLS metadata

```python
import json
import glob
from pandas.io.json import json_normalize
import geopandas
df_list = []
df_smry_list = []

def season_classifier(row):
    if '01' in row["start_month"]:
        return "wet"
    else:
        return "dry"

for fn in glob.glob('/projects/test_data/sajama/20230512_19/master_*_HLS.json'):

    with open(fn) as f:
        #data = pd.read_json(f)
        df = pd.json_normalize(json.load(f), record_path=['features'])
        df['filename'] = os.path.basename(fn)
        df['start'] = os.path.basename(fn).split('_')[2]
        df['start_month'] = os.path.basename(fn).split('_')[2].split('-')[1]
        df['year'] = os.path.basename(fn).split('_')[2].split('-')[0]
        df['season'] = 'none'
        df['season'] = df.apply(season_classifier, axis=1)
        df['end'] = os.path.basename(fn).split('_')[3]
        df_smry = df.groupby(['filename','year','season']).size().reset_index(name='counts')
        df_list.append(df)
        df_smry_list.append(df_smry)

df = pd.concat(df_list)
df_smry = pd.concat(df_smry_list)
df_smry.head()
```

#### 3.1.2_dps.py

```python
MS_DATA_TYPE = 'HLS'#'LS8' # 'HLS'
if MS_DATA_TYPE == 'LS8':
    SAT_API = 'https://landsatlook.usgs.gov/sat-api'
else:
    SAT_API = 'https://cmr.earthdata.nasa.gov/stac/LPCLOUD'   
```

```python
import fiona
from CovariateUtils import *
```

```python
nowtime = pd.Timestamp.now().strftime('%Y%m%d_%H')
```

```python
INDEX_FN = PROJECTED_FN

INDEX_LYR = fiona.listlayers(INDEX_FN)[0]
OUTDIR_SAJAMA = f'/projects/test_data/sajama/{nowtime}'
!mkdir -p $OUTDIR_SAJAMA

for SEASON in ['DRY','WET']:
    
    print(f'\n{SEASON}')
    
    if SEASON is 'WET':
        SEASON_START = '01-01' #'01-01'
        SEASON_STOP = '03-31' #'04-30'
        MAX_CLOUDS = 90
    else:
        SEASON_START = '06-01'
        SEASON_STOP = '09-30'
        MAX_CLOUDS = 25
         
    for YEAR in list(range(2013,2022+1)):

        YEAR = str(YEAR)
        
        args = f"-i {INDEX_FN} \
        -lyr {INDEX_LYR} \
        -a {SAT_API} \
        --tile_buffer_m 0 \
        --in_tile_num 1 \
        -o {OUTDIR_SAJAMA} \
        -sy {YEAR} -ey {YEAR} -smd {SEASON_START} -emd {SEASON_STOP} -mc {MAX_CLOUDS} \
        --composite_type {MS_DATA_TYPE} \
        --hls_product H30 \
        --rangelims_red 0 1 \
        --thresh_min_ndvi -1"
        #--search_only
        #"

        cmd = f'python /projects/Developer/icesat2_boreal/lib/3.1.2_dps.py {args}'
        !echo $cmd
        !echo
        !eval $cmd
```
